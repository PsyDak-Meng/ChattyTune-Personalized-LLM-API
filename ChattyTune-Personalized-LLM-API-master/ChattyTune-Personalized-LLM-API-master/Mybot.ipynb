## Load Pretrained model
Using GPT-neo-x-20B! Reducing model size to around 40GB in half precision to fit on ColabPro environment.

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import accelerate, bitsandbytes

# bit & bytes configuration
model_id = "EleutherAI/gpt-neox-20b"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config, device_map={"":0})
```

Apply some preprocessing to the model `prepare_model_for_kbit_training` method from PEFT.

```python
from peft import prepare_model_for_kbit_training

# selectively recomputing intermediate activations
model.gradient_checkpointing_enable()
model = prepare_model_for_kbit_training(model)
```

```python
def print_trainable_parameters(model):
    """
    Prints the number of trainable parameters in the model.
    """
    trainable_params = 0
    all_param = 0
    for _, param in model.named_parameters():
        all_param += param.numel()
        if param.requires_grad:
            trainable_params += param.numel()
    print(
        f"trainable params: {trainable_params} || all params: {all_param} || trainable%: {100 * trainable_params / all_param}"
    )
```

Reducing trainable parameter % to 0.08%!

```python
from peft import LoraConfig, get_peft_model

config = LoraConfig(
    r=8,
    lora_alpha=32,
    target_modules=["query_key_value"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)
config.inference_mode = False

model = get_peft_model(model, config)
print_trainable_parameters(model)
```

## Load Dataset
Chosen arxiv-cs-ml dataset.

Let's load a common dataset, english quotes, to fine tune our model on famous quotes.

```python
from datasets import load_dataset

dataset = load_dataset("ArtifactAI/arxiv-cs-ml-instruct-tune-50k")

# Define a function to process the data for training
def preprocess_function(examples):
    n = len(examples.keys())
    inputs = examples["question"]
    model_inputs = tokenizer(inputs, max_length=512, truncation=True)
    # Prepare labels
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples["answer"], max_length=512, truncation=True)
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Process the dataset
tokenized_dataset = dataset.map(preprocess_function, batched=True)
# dataset = data.map(lambda samples: tokenizer(samples["answer"]), batched=True)
```

## Training Step

```python
import transformers

# needed for gpt-neo-x tokenizer
tokenizer.pad_token = tokenizer.eos_token

trainer = transformers.Trainer(
    model=model,
    train_dataset=tokenized_dataset["train"],
    args=transformers.TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        warmup_steps=2,
        max_steps=50,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=1,
        output_dir="outputs",
        optim="paged_adamw_8bit",
    ),
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False),
)
model.config.use_cache = False  # silence the warnings. Please re-enable for inference!
trainer.train()
```

```python
# importing required modules
import PyPDF2

pdf_path = 'rag.pdf'

def read_pdf(pdf_path):
  summary = ''
  # creating a pdf file object
  pdfFileObj = open(pdf_path, 'rb')

  # creating a pdf reader object
  pdfReader = PyPDF2.PdfReader(pdfFileObj)

  n = (len(pdfReader.pages))
  title = pdf_path.split('/')[0]

  for i in range(n):
    # creating a page object
    pageObj = pdfReader.pages[0]

    # extracting text from page
    summary += pageObj.extract_text()

    # closing the pdf file object
    pdfFileObj.close()
  return summary, title

summary, title = read_pdf(pdf_path)
```

## Retrieval

```python
from transformers import AutoTokenizer
import numpy as np
from numpy.linalg import norm
from tqdm import tqdm


# This class holds a document in a form that the Chroma database likes.
# Namely, there's page_content with the
# actual payload you want to index in the database,
# and metadata--here, just a title
class RagRetriever:
  def __init__(self,summary, title, query):
    self.page_content = summary
    self.contents = None
    self.metadata = {"title":title}
    self.batch_sentences = []
    self.tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-cased")
    self.query = query
    self.similarities = None

  def cosine_similarity(self,A,B):
    cosine = np.dot(A,B)/(norm(A, axis=1)*norm(B))
    return cosine

  def batch_encode(self) -> None:
    self.contents = list(set(self.page_content.split('.')))
    n = int(len(self.contents)/128) + 1
    total = len(self.contents)
    self.similarities = similarities = np.zeros((total,))
    encoded_query = self.tokenizer(self.query, padding="max_length", max_length=128)['input_ids']

    for i in tqdm(range(n)):
      # Encode batch_size=128
      encoded_doc = []
      n = i+128 if (i+128) <total else total
      batch_sentences = self.contents[i:n]
      encoded_doc  = encoded_doc + self.tokenizer(batch_sentences, padding="max_length", max_length=128)['input_ids']
      self.similarities[i:n] = self.cosine_similarity(np.array(encoded_doc), np.array(encoded_query))

  def retrieve(self,k) -> str:
    """Retreive top k relevant non-repeating  sentences as RAG retrieval."""

    top_idx = self.similarities.argsort()[::-1]
    rag_retrieval = ''

    i = 0
    while i < k:
      if self.contents[top_idx[i]] not in rag_retrieval:
        retrieval = self.contents[top_idx[i]].replace('\n','')
        rag_retrieval += retrieval+'\n'
        i += 1
        print(i, retrieval)
    return rag_retrieval

```

```python
rag_retreiver = RagRetriever(summary, title, query)
rag_retreiver.batch_encode()
rag_retrieval = rag_retreiver.retrieve(10)
```

## Generation with RAG

```python
class RagGenerastor:
  def __init__(self, query, rag_retrieval, model) -> None:
    self.query = query
    self.rag_retrieval = rag_retrieval
    self.model = model
    self.template = f'You are my personal assistant,answer this {self.query} \n\
                      Reference the meta data and give me your own answer.\n\
                      Meta data: {self.rag_retrieval}\n\
                      Answer below:'

  def generate(self, max_tokens):
    inputs = tokenizer(self.template, return_tensors="pt").to(device)
    outputs = model.generate(**inputs, max_new_tokens=max_tokens)
    print(tokenizer.decode(outputs[0], skip_special_tokens=True))

rag_generator = RagGenerastor(query, rag_retrieval, model)
rag_generator.generate(50)
```
